```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv(r'E:\DSBDAL\DSBDALExam DataSets\DSBDALExam DataSets\HeartDisease\heart.csv')

df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x='target', y='chol', data=df, jitter=True)
plt.title("Cholesterol Levels vs Heart Disease")
plt.xlabel("Heart Disease (1 = Yes)")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='thalach', data=df)
plt.title("Average Max Heart Rate by Gender")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
df_sorted = df.sort_values('age')
plt.figure(figsize=(10, 5))
plt.plot(df_sorted['age'], df_sorted['thalach'], marker='o')
plt.title("Max Heart Rate by Age")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
# Box plot with histogram (using seaborn's displot)
sns.displot(df['chol'], kind='kde')  # histogram + curve
plt.title("Cholesterol Distribution with KDE")
plt.show()

# Scatter + Box
sns.boxplot(x='target', y='thalach', data=df)
sns.stripplot(x='target', y='thalach', data=df, color='red', jitter=True)
plt.title("Heart Disease vs Max Heart Rate with Individual Points")
plt.show()
```
