```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'E:\DSBDAL\DSBDALExam DataSets\DSBDALExam DataSets\HeartDisease\heart.csv', header=None)

# adding columns
data.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']

data.head()
```

```python
data = data.replace('?', np.nan)
data
```

## **1. Histogram**
**Objective: Visualize the distribution of ages in the dataset**

```python
plt.figure(figsize=(12, 10))
plt.hist(data['age'], bins=10, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution', size=20)
plt.show()
```

## **2. Pie Charts**
**Objective: Compare the distribution of sexes in the dataset**

```python
plt.figure(figsize=(12, 10))
data['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'red'])
plt.ylabel('')
plt.title('Sex Distribution', size=20)
plt.show()
```

## **3. Box Plots**
**Objective: Compare cholestrol levels for different chest pain types**

```python
plt.figure(figsize=(12, 10))
sns.boxplot(x='cp', y='chol', data = data)
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholestrol')
plt.title('Chest Pain vs. Cholestrol', size=20)
plt.show()
```

## **4. Scatter Plots**
**Objective: Explore the relationship between age and heart rate**

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(x='age', y='thalach', data=data, hue='num')
plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.title('Age vs. Heart Rate', size=20)
plt.show()
```

## **5. Bar Plot**
**Compare the frequency of different types of heart disease**

```python
plt.figure(figsize=(12, 10))
data['num'].value_counts().plot(kind='bar')
plt.xlabel('Heart Disease')
plt.ylabel('Frequency')
plt.title('Heart Disease Distribution', size=20)
plt.show()
```
